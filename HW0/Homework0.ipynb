```python
#imports
from matplotlib import pyplot as plt
import numpy as np
import scipy.optimize as opt

#reading data from file
data = open('fakeData.txt', 'r')

#reading in the first line in the data (x, y, bad)
header1 = data.readline()

#creating lists
good_data_x = []
good_data_y = []
bad_data_x = []
bad_data_y = []

#the last two lists are created to use for polynomial fitting
data_x = []
data_y = []

for line in data:
    line = line.strip() 
    line = line.split()
    columns = line #created a list of strings
    
    
    first_column = float(columns[0].replace(',', '')) #removing the commas and making the strings into floats
    second_column = float(columns[1].replace(',', ''))
    third_column = float(columns[2])
    
    #the following two are later used for polynomial fitting. This stores the unseperated data (good and bad together)
    data_x.append(first_column)
    data_y.append(second_column)
    
    #seperating bad data and good data coordinates into four different lists
    if third_column == -1:
        bad_data_x.append(first_column)
        bad_data_y.append(second_column)
    else:
        good_data_x.append(first_column)
        good_data_y.append(second_column)


print(6)
```

```python
#scatter plot of the good data
plt.scatter(good_data_x, good_data_y, color='blue', label='Good Data')

#scatter plot of the bad data
plt.scatter(bad_data_x, bad_data_y, marker='x', color='red', label='Bad Data')

#labels for the graph
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Good and bad data with polynomial fit')


polyfit2 = np.polyfit(data_x, data_y, 2) #stores the coeffcients of a degree 2 polynomial
equation = np.poly1d(polyfit2) #created the resulting polynomial with the coefficients from 'polyfit2'


x_new = np.arange(min(data_x), max(data_x)) #evenly spaed x-cooridnates for the fitted curve
y_fit = equation(x_new) #corresponding y-coordinates for the fitted curve

plt.plot(x_new, y_fit, color='orange', linewidth=3) #plotting the fit

plt.show()
```
